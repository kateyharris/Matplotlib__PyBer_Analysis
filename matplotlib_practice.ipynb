```python
# Create a Line Chart Using the MATLAB Method

%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot MATLAB Method
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() object-oriented approach
# create a figure
fig, ax = plt.subplots()
# define the axes
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() object-oriented approach
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()object-oriented approach
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the line plot MATLAB Method
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot MATLAB Method
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Import new dependency
import statistics
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Add Error Bars to a Line Chart
plt.errorbar(x_axis, y_axis, color="blue", linewidth=2, label='Boston', yerr=stdev, capsize=3)
plt.show()
```

```python
# Using the object-oriented approach, make the following changes to your line chart:
fig = plt.figure()
ax = fig.add_subplot()
# Change the line color to green.
# Change the marker to a diamond.
# Change the line width to 2 points.
ax.plot(x_axis, y_axis, marker="*", color="green", linewidth=2)
# Add a legend for the city of Boston.
ax.legend("Boston")

# Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Add a y-axis limit.
ax.set_ylim(10)
# Add grid lines.
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#We will make the size of each marker equal to the data floating-point decimal values of the y-axis
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# adjust the size by multiplying the data in the y-axis by 3
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
# use the new y-axis list for the s parameter
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# refactor the code we used to create a scatter plot above, 
# using Python's list comprehension technique inside the plt.scatter() function.
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Using the object-oriented approach, make the following changes to your bubble chart:
fig, ax = plt.subplots()
# Change the color of the markers to sky blue, Make the color 20% transparent.
# Change the size of the markers to 5 times each data point.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.
#ax.scatter(x_axis, y_axis, s=[i * 5 for i in y_axis], facecolor="lightblue", alpha=0.2, linewidth=2, edgecolor='black')

# Add a legend for the city of Boston.
ax.legend(labels="", title="Boston")
# Add a title.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Dates")
ax.set_ylabel("Fare($)")
# Switch the axis so that the Fare($) data is on the x-axis.
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color="skyblue", alpha=0.8, linewidth=2, edgecolor='black')

# Add a limit for the x-axis data.
ax.set_xlim(10)
# Invert the y-axis so that January is at the top.
ax.invert_yaxis()
```

```python
# add percentages for each month and "explode" the largest percentage, which is July, the seventh value in the x_axis
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Using the object-oriented approach, make the following changes to the pie chart:
# Increase the figure size to 8x8.
fig, ax = plt.subplots(figsize=(8, 8))

# Add new colors of your own choosing or use the colors from the previous pie chart.
colors = ["slategrey", "lightsteelblue", "cornflowerblue", "royalblue", "midnightblue", "indigo", "darkorchid", "blueviolet", "darkmagenta", "purple", "plum", "thistle"]
# Explode the two highest percentage months (3rd and 7th).
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis, colors=colors, 
       explode=explode_values, 
       labels=x_axis, 
# Add a percentage to one decimal place to each wedge of the pie.       
       autopct='%.1f%%',
# Add a shadow.
      shadow=True,
# Add a start angle so that January is at the top.
      startangle=90,
# Reverse the order so that the month order is in a clockwise direction.
      counterclock=False)
plt.show()
```
